```python
import pandas as pd 
import numpy as np 
```

## **Adj matrix from Barbera 1**

```python
adj_2 = pd.read_csv('data/misc/Replication_data_Barbera_2014/adj-matrix-US.csv')
barbera_adj_elites = [i.lower() for i in list(adj_2.columns[1:])]
barbera_adj_elites = pd.DataFrame(barbera_adj_elites).rename(columns = {0:'twitter_name'})
print(f'{len(barbera_adj_elites)} elites in adj')
```

**Pivot into edgelist**

```python
network_edgelist = adj_2.melt(id_vars=['Unnamed: 0'], var_name='twitter_name', value_name='follow')
network_edgelist = network_edgelist[network_edgelist['follow'] != 0].rename(columns={'Unnamed: 0' : 'user'}).drop(['follow'],axis=1)
network_edgelist['twitter_name'] = network_edgelist['twitter_name'].str.lower()
print(f'{network_edgelist['user'].nunique()} users')
print(f'{network_edgelist['twitter_name'].nunique()} twitter_name')
```

**Filter for 183 congressmembers**

```python
zhixiang_house = pd.read_csv('data/misc/Data_about_Opinion_Measurement_Zhixiang/1. Following Data/2. Information about congressional members insde 318 accounts/House_list_in_318_accounts.csv')
zhixiang_senate = pd.read_csv('data/misc/Data_about_Opinion_Measurement_Zhixiang/1. Following Data/2. Information about congressional members insde 318 accounts/Senate_list_in_318_accounts.csv')

zhixiang_congress = pd.concat([zhixiang_house, zhixiang_senate]).drop_duplicates(subset=['twitter_name']).drop(['DW-Nominate'], axis=1)

zhixiang_results_house = pd.read_csv('data/misc/Data_about_Opinion_Measurement_Zhixiang/2.Opinion results/opinion_house.csv')
zhixiang_results_senate = pd.read_csv('data/misc/Data_about_Opinion_Measurement_Zhixiang/2.Opinion results/opinion_senate.csv')

zhixiang_results_congress = pd.concat([zhixiang_results_house, zhixiang_results_senate]).drop_duplicates(subset=['twitter_name'])

zhixiang_list = zhixiang_congress.merge(zhixiang_results_congress, on='twitter_name', how='inner')
zhixiang_list
```

```python
network_edgelist_183 = network_edgelist.merge(zhixiang_list[['twitter_name','chamber']], on='twitter_name', how='inner')
```

```python
network_edgelist_183
```

```python
network_edgelist_183_house = network_edgelist_183[network_edgelist_183['chamber'] == 'House'][['user', 'twitter_name']]
network_edgelist_183_house.to_csv('data/misc/Data_about_Opinion_Measurement_Zhixiang/network_edgelist_house.csv', header=None, index=None)
```

```python
network_edgelist_183_senate = network_edgelist_183[network_edgelist_183['chamber'] == 'Senate'][['user', 'twitter_name']]
network_edgelist_183_senate.to_csv('data/misc/Data_about_Opinion_Measurement_Zhixiang/network_edgelist_senate.csv', header=None, index=None)
```
